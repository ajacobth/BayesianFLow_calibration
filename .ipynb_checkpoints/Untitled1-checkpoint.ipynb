```python
from __future__ import annotations

from pathlib import Path
from math import ceil, log2
from typing import Optional, Union, Sequence, List, Dict, Tuple

import numpy as np
import pandas as pd
from scipy.stats import qmc, lognorm, norm


ArrayLike = Union[float, int, Sequence[float], np.ndarray]


class Reflectance_Surrogate:
    """
    Surrogate that:
      - loads (wavelength, n, k) from an Excel file
      - provides Sobol train/test inputs (exact N points)
      - interpolates n, k at arbitrary wavelength(s)
      - assembles forward inputs for an external reflectance model

    Notes:
      * Units for wavelength must match your Excel file (e.g., meters or nm).
      * f_min == f_max == 1.0 below → f is fixed at 1.0 (adjust if needed).
      * much of the doucemntation was written by chatGPT
    """

    def __init__(
        self,
        dim: int,
        n_train_samples: int,
        n_test_samples: int,
        *,
        excel_path: Optional[str] = None,  # path to n,k table
        sheet_name: Union[int, str] = 0,   # sheet index/name
        wavelength_col: Optional[str] = None,
        n_col: Optional[str] = None,
        k_col: Optional[str] = None,
        allow_extrapolation: bool = False,  # False → NaN out of range
        # kept for future use (not used below):
        bounds: Optional[np.ndarray] = None,
        segments: Optional[List[Dict]] = None,
    ):
        self.dim = int(dim)
        self.n_train_samples = int(n_train_samples)
        self.n_test_samples = int(n_test_samples)
        self.allow_extrapolation = bool(allow_extrapolation)

        self._wl: Optional[np.ndarray] = None
        self._n: Optional[np.ndarray] = None
        self._k: Optional[np.ndarray] = None
        self._wl_min: Optional[float] = None
        self._wl_max: Optional[float] = None

        # -------- parameter bounds (tune as needed) --------
        # Parameters: λ, f, n, k, r, h, L, μ0
        self.lam_max = 3e-5  # got this from the excel file
        self.lam_min = 5e-7  # got this from the excel file

        # NOTE: set both to 1.0 → f will be fixed at 1.0
        self.f_max = 0.6
        self.f_min = 0.01

        self.n_max = None  # not used; n comes from interpolated mean + noise
        self.n_min = None
        self.k_max = None
        self.k_min = None

        self.lg_shape_max = 0.6  # lg is log normal distribution
        self.lg_shape_min = 0.1

        self.lg_scale_max = 5.0  # scale of lognormal
        self.lg_scale_min = 0.1

        self.L_max = 5e-3
        self.L_min = 1e-3

        self.mu0_max = 1.0
        self.mu0_min = 0.79
        # ---------------------------------------------------

        if excel_path is not None:
            df = self._load_optical_constants_excel(
                excel_path, sheet_name, wavelength_col, n_col, k_col
            )
            self._set_nk_from_df(df, "wavelength", "n", "k")
        else:
            raise ValueError("Provide excel_path to load (wavelength, n, k).")

    # -------------------- Excel Loader -------------------- #
    def _load_optical_constants_excel(
        self,
        path: str,
        sheet_name: Union[int, str] = 0,
        wavelength_col: Optional[str] = None,  # kept for signature compatibility; ignored
        n_col: Optional[str] = None,           # ignored
        k_col: Optional[str] = None,           # ignored
    ) -> pd.DataFrame:
        
        """Currently only excel files, but"""
        p = Path(path)
        if not p.is_file():
            raise FileNotFoundError(f"Excel file not found: {p}")

        ext = p.suffix.lower()
        if ext in (".xlsx", ".xlsm", ".xltx", ".xltm"):
            engine = "openpyxl"
        elif ext == ".xls":
            engine = "xlrd"
        else:
            raise ValueError(f"Unsupported file extension: {ext}")

        try:
            # Read the sheet; we will select by column POSITION, not name
            df_raw = pd.read_excel(p, sheet_name=sheet_name, engine=engine)
        except ImportError as e:
            need = "openpyxl" if engine == "openpyxl" else "xlrd"
            raise ImportError(
                f"Missing Excel reader '{need}'. Install via `pip install {need}`. Original error: {e}"
            ) from e

        # Ensure we have at least 3 columns  - else something is missing
        if df_raw.shape[1] < 3:
            raise ValueError(
                f"Expected at least 3 columns (wavelength, n, k) but found {df_raw.shape[1]}."
            )

        # Take the first three columns as wavelength, n, k (position-based)
        df = df_raw.iloc[:, :3].copy()
        c0, c1, c2 = df.columns[:3]
        df = df.rename(columns={c0: "wavelength", c1: "n", c2: "k"})

        # Clean & validate
        df = df.dropna(subset=["wavelength", "n", "k"])
        df["wavelength"] = pd.to_numeric(df["wavelength"], errors="coerce")
        df["n"] = pd.to_numeric(df["n"], errors="coerce")
        df["k"] = pd.to_numeric(df["k"], errors="coerce")
        df = df.dropna(subset=["wavelength", "n", "k"])

        # Combine duplicate wavelengths by averaging
        df = df.groupby("wavelength", as_index=False)[["n", "k"]].mean()

        # Sort
        df = df.sort_values("wavelength").reset_index(drop=True)

        if df.empty:
            raise ValueError("No valid (wavelength, n, k) rows after cleaning the Excel sheet.")

        return df

    def _set_nk_from_df(self, df: pd.DataFrame, wl_col: str, n_col: str, k_col: str) -> None:
        self.set_nk_arrays(df[wl_col].to_numpy(), df[n_col].to_numpy(), df[k_col].to_numpy())

    def set_nk_arrays(self, wavelength: ArrayLike, n: ArrayLike, k: ArrayLike) -> None:
        wl = np.asarray(wavelength, dtype=float)
        n = np.asarray(n, dtype=float)
        k = np.asarray(k, dtype=float)
        if not (wl.shape == n.shape == k.shape):
            raise ValueError("wavelength, n, k must have the same shape.")
        order = np.argsort(wl)
        wl = wl[order]
        n = n[order]
        k = k[order]
        wl_unique, idx = np.unique(wl, return_index=True)
        n_unique = n[idx]
        k_unique = k[idx]
        self._wl = wl_unique
        self._n = n_unique
        self._k = k_unique
        self._wl_min = float(self._wl[0])
        self._wl_max = float(self._wl[-1])

    # ---------------- Linear Interpolation for n,k ---------------- #
    def interpolate_k_n(self, lamb: ArrayLike) -> np.ndarray:
        """
        Return interpolated [[n],[k]] for wavelength(s) `lamb`.
        - scalar -> shape (2,1)
        - array  -> shape (2,M)
        """
        if self._wl is None:
            raise RuntimeError(
                "No (wavelength, n, k) data loaded. Load from Excel first."
            )

        xq = np.atleast_1d(np.asarray(lamb, dtype=float))

        def _interp(y: np.ndarray) -> np.ndarray:
            if self.allow_extrapolation:
                return np.interp(xq, self._wl, y)
            yq = np.interp(np.clip(xq, self._wl_min, self._wl_max), self._wl, y)
            oob = (xq < self._wl_min) | (xq > self._wl_max)
            yq[oob] = np.nan
            return yq

        n_q = _interp(self._n)
        k_q = _interp(self._k)
        out = np.vstack([n_q, k_q])
        if np.isscalar(lamb):
            return out.reshape(2, 1)
        return out

    # ---------------- Sobol helpers / datasets ---------------- #
    def _sobol_exact(self, n: int, d: int, scramble: bool = True) -> np.ndarray:
        """Return exactly `n` Sobol points in [0,1]^d."""
        if n <= 0:
            raise ValueError("n must be positive.")
        sampler = qmc.Sobol(d=d, scramble=scramble)
        m = int(ceil(log2(n)))
        X = sampler.random_base2(m=m)  # 2**m points
        return X[:n]

    def create_training_set(self, dim: int, n: int, override: bool = False, slicing: bool = False) -> np.ndarray:
        """
        override=True  -> treat `n` as exponent m, return 2**n points
        else            -> return exactly `self.n_train_samples` points
        If slicing=True and not override, return exactly `n` points.
        """
        if override:
            sampler = qmc.Sobol(d=dim, scramble=True)
            return sampler.random_base2(m=n)
        if slicing:
            return self._sobol_exact(n, dim, scramble=True)
        return self._sobol_exact(self.n_train_samples, dim, scramble=True)

    def create_test_set(self, dim: int, n: int, override: bool = False, slicing: bool = False) -> np.ndarray:
        if override:
            sampler = qmc.Sobol(d=dim, scramble=True)
            return sampler.random_base2(m=n)
        if slicing:
            return self._sobol_exact(n, dim, scramble=True)
        return self._sobol_exact(self.n_test_samples, dim, scramble=True)

    # ---------------- Lognormal binning ---------------- #
    def lognorm_bin_probs(
        self,
        s: float,
        scale: float,
        loc: float = 0.0,
        bins: Union[str, int, Sequence[float]] = "fd",
        n_mc: int = 1000,
        rng: Optional[np.random.Generator] = None,
    ) -> Tuple[np.ndarray, np.ndarray]:
        """
        Return (centers, probs) for a lognormal defined by (s, scale, loc).
        - centers: bin centers (vector r)
        - probs: probability per bin (vector h), sums to 1
        """
        rng = rng or np.random.default_rng()
        rv = lognorm(s=s, loc=loc, scale=scale)

        # Use MC draws to choose reasonable edges; probabilities from analytic CDF
        data = rv.rvs(size=n_mc, random_state=rng)
        edges = np.histogram_bin_edges(data, bins=bins)

        probs = np.diff(rv.cdf(edges))            # probability per bin
        centers = 0.5 * (edges[:-1] + edges[1:])  # bin centers

        ssum = probs.sum()
        if ssum > 0:
            probs = probs / ssum
        return centers, probs

    # ---------------- Forward model glue ---------------- #
    def forward_reflectance(self, lam, f, n, k, r, h, L, mu0) -> float:
        """
        Placeholder for Sergio's reflectance model.
        r, h are vectors (same length); others are scalars.
        """
        # TODO: replace with real model call here for Sergios function
        return 0.0

    def _forward(self, z1, z2, z3, z4, z5, z6, z7, z8) -> float:
        """
        z1: λ (scaled 0-1)
        z2: f (scaled 0-1)
        z3: noise for n (unit uniform; transformed via Φ^{-1})
        z4: noise for k (unit uniform; transformed via Φ^{-1})
        z5: lognormal shape (scaled)
        z6: lognormal scale (scaled)
        z7: fiber length (scaled)
        z8: μ0 (scaled)
        """
        lam = z1 * (self.lam_max - self.lam_min) + self.lam_min
        f   = z2 * (self.f_max  - self.f_min)   + self.f_min

        nk = self.interpolate_k_n(lam)  # (2,1) for scalar λ
        n_mean = float(nk[0, 0])
        k_mean = float(nk[1, 0])

        z3_inv = norm.ppf(z3)
        z4_inv = norm.ppf(z4)
        n = n_mean * (1.0 + 0.075 * z3_inv)
        k = k_mean * (1.0 + 0.075 * z4_inv)

        lg_shape = z5 * (self.lg_shape_max - self.lg_shape_min) + self.lg_shape_min
        lg_scale = z6 * (self.lg_scale_max - self.lg_scale_min) + self.lg_scale_min

        r_vec, h_vec = self.lognorm_bin_probs(lg_shape, lg_scale, loc=0.0, bins="fd")

        L   = z7 * (self.L_max - self.L_min) + self.L_min
        mu0 = z8 * (self.mu0_max - self.mu0_min) + self.mu0_min

        ref = self.forward_reflectance(lam, f, n, k, r_vec, h_vec, L, mu0)
        return ref



```

```python
shape, loc, scale = stats.lognorm.fit(radius_s, floc=0)
print(shape, scale)
# Create the fitted distribution object
fitted_rv = stats.lognorm(s=shape, loc=loc, scale=scale)

# --- Compare data vs fitted PDF ---
# common x range: from min(data) to ~99th percentile to avoid long-tail distortion
x = np.linspace(0, np.percentile(radius_s, 99.5), 1000)

plt.figure(figsize=(8,5))

# Histogram of data normalized to density
plt.hist(radius_s,bins="fd", density=True, alpha=0.5, color='red', label='Data')

# Fitted PDF (smooth curve)
plt.plot(x, fitted_rv.pdf(x), 'b-', lw=2, label='Fitted lognormal PDF')

plt.title("Data vs Fitted Lognormal")
plt.xlabel("x")
plt.ylabel("Density")
plt.legend()
plt.show()

print("Fitted distirbution mean is :", scale * np.exp(0.5 * shape**2))
print("Sample mean is:", np.mean(radius_s))
```

```python
def lognorm_bin_probs(s, scale, loc=0.0,bins='fd'):
    """
    Return bin centers and probabilities for a fitted lognormal.
    """
    rv = lognorm(s=s, loc=loc, scale=scale)
    data = fitted_rv.rvs(1000)

    edges = np.histogram_bin_edges(data, bins=bins)

    probs = np.diff(rv.cdf(edges)) # probability of each bin

    centers = 0.5 * (edges[:-1] + edges[1:])  # centers of the bin

    return centers, probs
```
